```python
import numpy as np
df = pd.read_csv("o1.csv")
def convert_k(val):
    if pd.isna(val):
        return 0
    val = str(val).lower().strip()
    if val.endswith('k'):
        return int(float(val.replace('k', '')) * 1000)
    return int(float(val))

df['Times_Listed'] = df['Times_Listed'].apply(convert_k)
```

```python
import pandas as pd

def clean_reviews(val):
    if pd.isna(val):
        return 0
    val = str(val).lower().strip()

    if val.endswith('k'):
        return int(float(val.replace('k', '')) * 1000)

    if val.endswith('m'):
        return int(float(val.replace('m', '')) * 1_000_000)

    val = val.replace(',', '')
    return int(float(val)) if val.replace('.', '').isdigit() else 0

df['Number_of_Reviews'] = df['Number_of_Reviews'].apply(clean_reviews)
```

```python
print(df['Number_of_Reviews'].head(10))
print(df['Number_of_Reviews'].dtype)
print(df['Number_of_Reviews'].max())
```
